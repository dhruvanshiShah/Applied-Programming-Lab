```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
 # Main system parameters: number of mics, number of samples in time 
Nmics = 64
Nsamp = 200

# Source: x,y coordinates: x: 0+, y: [-Y, +Y] where Y determined by pitch and Nmics 
src = (0, 0)
pitch = 0.1
# Spacing between microphones pitch = 0.1 proxy for sampling rate
dist_per_samp = 0.1
 # Speed of sound in the medium
C = 2
 # Time dilation factor for sinc pulse: how narrow
SincP = 5.0

 # CODE Locations of microphones
mics = []
i = 0
while(i < Nmics):
    if i != Nmics//2:
        mics.append([0, round(i*pitch - (Nmics * pitch / 2), 2)])
    else:
        if Nmics%2 == 0:
            mics.append([0, round((i + 1)*pitch - Nmics * pitch / 2, 2)])
            i += 2
            continue
        else:
            mics.append([0, 0])
    if i == Nmics - 1 and Nmics%2 == 0:
        mics.append([0, round((i + 1)*pitch - (Nmics * pitch / 2), 2)])
    i += 1
 # Location of point obstacle
obstacle = (3, -1)                  # for generating intensity(output waveform of mics)
x_start, y_start = int(mics[0][0]/(1)), int(mics[0][1]/(1))
x_end, y_end = int(mics[len(mics) - 1][0]/(1)), int(mics[len(mics) - 1][1]/(1))
```

```python
def distance(a, b):
    return ((b[0] - a[0])**2 + (b[1] - a[1])**2)**0.5

# Distance from src to a mic after reflecting through pt
def dist(src, pt, mic):
    d1 = distance(src, pt) # CODE distance from src to pt
    d2 = distance(pt, mic) # CODE distance from pt to mic
    return d1 + d2
```

```python
# Source sound wave - time axis and wave
# sinc wave with narrowness determined by parameter
t = 0 # CODE Nsamp time instants with spacing of dist_per_samp
# SincP = 0.1
def wsrc(t):                        
        return  np.sinc(SincP*(t))
```

```python
t = np.arange(0, 20, 0.1)
intensity = np.zeros((len(mics), 200))
for i_mic in range(len(mics)):
    total_dist = dist(src, (3, -1), mics[i_mic])
    delay = total_dist/C
    for t0 in t:
        if t0 - delay < 0:
            intensity[i_mic][int(t0/0.1)] = 0
        else:
            intensity[i_mic][int(t0/0.1)] = wsrc(t0 - delay)

amp_mic = np.zeros((64, 200))

for y in range(-32, 32):
    for x in range(200):
        amp = 0
        for i_mic in range(len(mics)):
            total_dist = dist(src, (x*dist_per_samp, y*dist_per_samp), mics[i_mic])
            delay = total_dist/C
            try: 
                amp += intensity[i_mic][int(delay/dist_per_samp) + 1]
            except Exception as e:
                amp += 0
        amp_mic[y+32][x] = amp
plt.imshow(amp_mic, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Heatmap for Dataset 1')
plt.savefig('dataset1.png')
```
